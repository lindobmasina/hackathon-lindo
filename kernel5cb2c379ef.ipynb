```python
#data11.head()
test['TransactionId'] = test['TransactionId'].map(lambda x: x.lstrip('TransactionId_'))
test['CustomerId'] = test['CustomerId'].map(lambda x: x.lstrip('CustomerId_'))
test['ProviderId'] = test['ProviderId'].map(lambda x: x.lstrip('ProviderId_'))
test['BatchId'] = test['BatchId'].map(lambda x: x.lstrip('BatchId_'))
test['AccountId'] = test['AccountId'].map(lambda x: x.lstrip('AccountId_'))
test['SubscriptionId'] = test['SubscriptionId'].map(lambda x: x.lstrip('SubscriptionId_'))
test['ProductId'] = test['ProductId'].map(lambda x: x.lstrip('ProductId_'))
test['ChannelId'] = test['ChannelId'].map(lambda x: x.lstrip('ChannelId_'))


test = test.drop(['CurrencyCode','TransactionStartTime'], axis=1)
#test['TransactionStartTime'] = test['TransactionStartTime'].apply(pd.to_datetime).map(lambda x: x.toordinal())
```

```python
#data11.head()
data11['TransactionId'] = data11['TransactionId'].map(lambda x: x.lstrip('TransactionId_'))
data11['CustomerId'] = data11['CustomerId'].map(lambda x: x.lstrip('CustomerId_'))
data11['ProviderId'] = data11['ProviderId'].map(lambda x: x.lstrip('ProviderId_'))
data11['BatchId'] = data11['BatchId'].map(lambda x: x.lstrip('BatchId_'))
data11['AccountId'] = data11['AccountId'].map(lambda x: x.lstrip('AccountId_'))
data11['SubscriptionId'] = data11['SubscriptionId'].map(lambda x: x.lstrip('SubscriptionId_'))
data11['ProductId'] = data11['ProductId'].map(lambda x: x.lstrip('ProductId_'))
data11['ChannelId'] = data11['ChannelId'].map(lambda x: x.lstrip('ChannelId_'))


data11 = data11.drop(['CurrencyCode','TransactionStartTime'], axis=1)
#data11['TransactionStartTime'] = data11['TransactionStartTime'].apply(pd.to_datetime).map(lambda x: x.toordinal())
```

```python
# The list comprehension below generates a list of columns whose entries are of the object data type
one_hot_list = ['ProductCategory']

# The following code implement one-hot encoding
data = pd.get_dummies(data11, columns = one_hot_list, prefix = one_hot_list, drop_first = True)
test = pd.get_dummies(test, columns = one_hot_list, prefix = one_hot_list, drop_first = True)

test.set_index('TransactionId')
data11.set_index('TransactionId')
```

```python
test.head()
```

```python
columns = data.columns.tolist()
# Filter the columns to remove data we do not want 
columns = [c for c in columns if c not in ["FraudResult"]]
# Store the variable we are predicting 
target = "FraudResult"
# Define a random state 
state = np.random.RandomState(42)
X = data[columns]
y = data[target]

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
sum(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lm = RandomForestClassifier()
lm.fit(X_train,y_train)
pred_lm = lm.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print('Classification Report')
print(classification_report(y_test, pred_lm))
```
